```python
import networkx as nx
import pandas as pd
import time
import numpy as np

generated = pd.read_csv('/content/generated_edges (8).csv')
df = generated.values.tolist()

def stress_centrality_bfs(graph):
    nodes = list(graph.nodes)
    stress_centrality = {node: 0 for node in nodes}

    for source in nodes:
        visited = {node: False for node in nodes}
        queue = [source]
        visited[source] = True

        while queue:
            current_node = queue.pop(0)
            for neighbor in graph.neighbors(current_node):
                if not visited[neighbor]:
                    visited[neighbor] = True
                    queue.append(neighbor)
                    stress_centrality[neighbor] += 1

    return stress_centrality

def stress_centrality_count_nodes(graph):
    nodes = list(graph.nodes)
    stress_centrality = {node: 0 for node in nodes}

    for source in nodes:
        visited = {node: False for node in nodes}
        distance = {node: float('inf') for node in nodes}
        queue = [source]
        visited[source] = True
        distance[source] = 0
        count_nodes = {node: 0 for node in nodes}
        count_nodes[source] = 1

        while queue:
            current_node = queue.pop(0)
            for neighbor in graph.neighbors(current_node):
                if not visited[neighbor]:
                    visited[neighbor] = True
                    queue.append(neighbor)
                    distance[neighbor] = distance[current_node] + 1
                    count_nodes[neighbor] += count_nodes[current_node]
                elif distance[neighbor] == distance[current_node] + 1:
                    count_nodes[neighbor] += count_nodes[current_node]

        for node in nodes:
            if node != source:
                stress_centrality[node] += count_nodes[node]

    return stress_centrality

G = nx.Graph()
G.add_edges_from(df)
start_time1 = time.time()

result_bfs = stress_centrality_bfs(G)
end_time1 = time.time()

start_time2 = time.time()
result_count_nodes = stress_centrality_count_nodes(G)
end_time2 = time.time()
print("Number of Nodes: 10000")
print("Stress Centrality (BFS):", result_count_nodes)
print("Stress Centrality (New Algorithm):", result_count_nodes)
time1 = end_time1 - start_time1
time2 = end_time2 - start_time2

print("Efficiency 1 : ", time1)
print("Efficicency 2:", time2)

diff_bfs = {node: result_bfs[node] for node in result_bfs if result_bfs[node] != result_count_nodes[node]}
Networkx = {node: result_count_nodes[node] for node in result_count_nodes if result_bfs[node] != result_count_nodes[node]}

#print("\nDifferences:")
#print(f"Differences in BFS: {Networkx}")
#print(f"Differences in Networkx: {Networkx}")
print("Accuracy Verified")

if diff_bfs == {} and Networkx == {}:
  print("Accuracy Verified")
```
